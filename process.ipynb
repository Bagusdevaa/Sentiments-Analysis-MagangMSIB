```python
import pandas as pd
import function
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# tweets 2024
df_2024 = pd.read_csv('data/magang.old.csv')
df1_2024 = pd.read_csv('data/magang.csv')

# tweets 2023
df_2023 = pd.read_csv('data/magang2023.csv')
df1_2023 = pd.read_csv('data/magang2023again.old.csv')
df2_2023 = pd.read_csv('data/magang2023again1.csv')

# tweets 2022
df_2022 = pd.read_csv('data/magang2022.csv')

# tweets 2021
df_2021 = pd.read_csv('data/magang2021.csv')
```

```python
# Menggabungkan semua data 
dfs = [df_2024, df1_2024, df_2023, df1_2023, df2_2023, df_2022, df_2021]

combined_df = pd.concat(dfs, ignore_index=True)
```

```python
# Mengecek seluruh data dari tahun 2021 s.d. 2024
t2024 = 0
t2023 = 0
t2022 = 0
t2021 = 0
for i in combined_df['created_at']:
    if '2024' in i:
        t2024 += 1
    elif '2023' in i:
        t2023 += 1
    elif '2022' in i:
        t2022 += 1
    elif '2021' in i:
        t2021 += 1

print('data tahun 2024: ', t2024)
print('data tahun 2023: ', t2023)
print('data tahun 2022: ', t2022)
print('data tahun 2021: ', t2021)
```

```python
# Daftar kolom yang ingin diambil
df = combined_df[['created_at','full_text', 'tweet_url', 'username']]
df
```

```python
# cek NaN
df.isnull().sum()
```

```python
# Cek duplikat
count = 0
for check in df['full_text'].duplicated():
    if check == True:
        count += 1
print(f'ada {count} banyak baris duplikat')
```

```python
# drop duplikat
row = 0
for check in df['full_text'].duplicated():
    if check == True:
        df = df.drop(index=row)
    row += 1
```

```python
# cek jumlah duplikat lagi
count = 0
for check in df['full_text'].duplicated():
    if check == True:
        count += 1
print(f'ada {count} banyak baris duplikat')
```
